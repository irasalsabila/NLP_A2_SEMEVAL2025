```python
import zipfile
import tarfile
import os

# File paths
files = [
    "dev-documents_25_October.zip",
    "semeval2025task10-scorers-baselines.tar",
    "training_data_16_October_release.zip"
]

# Directory for extracted files
extracted_dir = "extracted_files"
os.makedirs(extracted_dir, exist_ok=True)

# Define passwords for specific files
passwords = {
    "training_data_16_October_release.zip": "narratives5202trainTWO",
    "dev-documents_25_October.zip": "devNarratives5202"
}

# Function to extract ZIP files (with optional password)
def extract_zip_with_password(file_path, extract_to, password=None):
    with zipfile.ZipFile(file_path, 'r') as zip_ref:
        if password:
            zip_ref.setpassword(password.encode('utf-8'))
        zip_ref.extractall(extract_to)

# Function to extract TAR files
def extract_tar(file_path, extract_to):
    with tarfile.open(file_path, 'r') as tar_ref:
        tar_ref.extractall(extract_to)

# Extract files
for file in files:
    file_path = file
    extract_to = os.path.join(extracted_dir, os.path.splitext(file)[0])
    os.makedirs(extract_to, exist_ok=True)
    try:
        if file.endswith('.zip'):
            extract_zip_with_password(file_path, extract_to, passwords.get(file))
        elif file.endswith('.tar'):
            extract_tar(file_path, extract_to)
    except Exception as e:
        print(f"Failed to extract {file}: {e}")

# List contents of extracted files
extracted_contents = {}
for root, dirs, files in os.walk(extracted_dir):
    for name in dirs + files:
        extracted_contents.setdefault(root, []).append(name)

# Display the extracted contents
for folder, contents in extracted_contents.items():
    print(f"Folder: {folder}")
    for item in contents:
        print(f"  - {item}")
```
